#### CTD

The instruments are similar in that they both essentially measure three parameters: 
- The electrical **Conductivity** of seawater (***C***)
- The water **Temperature** (***T***)
- The **Pressure** (a proxy for depth, ***P***)

The measurement **Depth** (***D***) can be calculated from pressure\* - sets of measurements of these three parameters are often referred to as **CTD** measurements, and instruments measuring them are often referred to as **CTD**s.  

If we have ***C***, ***T*** and ***P***, we can calculate **Salinity** (***S***) using well-known formulas. CTD measurements are by far the most common method for measuring salinity in oceanography. 

In addition to salintiy, CTD measurements can be used to calculate the **Density** (***ρ***) and **Speed of sound** of the seawater - both of which are known functions of salinity, temperature, and pressure.  

In modern physical oceanography, we increasingly tend to use **Conservative Temperature** (***CT*** or ***Θ***) and **Absolute Salinity** (***SA***, ***$S_A$***) in analysis, rather than *practical salinity* and *in-situ* or *potential temperature*, which were more common in the past. CT and SA are thermodynamically consistent quantities that allow more consistent comparison of water masses collected at different depths (among other things). The conventions defining these quantities (including a very nice and accessible [primer](https://www.teos-10.org/pubs/TEOS-10_Primer.pdf) on the concepts) can be found here: [TEOS-10](https://www.teos-10.org/)

## Python libraries and Jupyter

The goal of this series is to illustrate a procedure for rudimentary analysis of data, i.e., the *steps* involved in basic post-processing and analysis. We will perform these operations in a certain way, using the Python language and publicly available Python libraries listed below. However, there are many ways to do and you may choose different sequence of events, using other libraries or programming languages, or coding things up yourself.    


##### ***Python libraries***
- `gsw` - Widely used toolbox for oceanographic operations.
- `numpy` - Standard python package for numerical operations.
- `xarray` - Extremely useful modern library for working with labelled data with one or multiple dimensions. 
- `matplotlib` - Vizualization library (creating nice looking plots). 
- `kval` - Custom library used to facilitate work with ocean data.

We will initially convert our input data to `xarray` `Dataset`s. This makes things like slicing data, up-and downscaling operations, and metadata handling relatively simple. Another advantage is that `xarray` `Dataset`s interface exellently to the NetCDF format - which is the community's preferred format for publication of datasets.
